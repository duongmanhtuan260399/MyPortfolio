## Report: act_report

This report presents the insights gained from analyzing the WeRateDogs Twitter data after the data wrangling process. The dataset was cleaned and analyzed to uncover interesting patterns and trends related to dog breeds, posting platforms, and the accuracy of an AI model in predicting dog breeds. Additionally, a visualization in the form of a boxplot was created to showcase the distribution of dog ratings.

### Insights:

1. Top 10 Most Common Dog Breeds: The analysis revealed that the most common dog breeds mentioned in the WeRateDogs tweets are Golden Retriever, Labrador Retriever, Pembroke, Chihuahua, Pug, Toy Poodle, Chow, Samoyed, Pomeranian, and Malamute. These breeds received the highest number of mentions in the dataset.

2. Preferred Posting Platform: The analysis indicated that people tend to prefer posting on their iPhones more than any other platform. iPhones were the most frequently used devices for posting tweets about dogs, suggesting a strong preference for mobile tweeting.

3. Accuracy of AI Model in Predicting Dog Breeds: The average precision of an AI model in predicting dog breeds was found to be 54.99%. While this percentage may not be considered impressive, it still provides a reliable means to identify dog breeds from images.

### Visualization: 

A boxplot was created to visualize the distribution of dog ratings. The boxplot showcases the range, median, and quartiles of the ratings given to dogs in the WeRateDogs tweets. This visualization allows for a quick understanding of the distribution and variability of dog ratings in the dataset.



The analysis of the WeRateDogs Twitter data revealed interesting insights about the most common dog breeds, preferred posting platforms, and the accuracy of an AI model in predicting dog breeds. These findings provide valuable information for dog enthusiasts, researchers, and social media analysts. The boxplot visualization enhances the understanding of the distribution of dog ratings. Overall, this analysis demonstrates the potential of data analysis to uncover insights and trends in social media data.
